```python
import lsdb
from dask.distributed import Client
```

```python
client = Client(n_workers=12, memory_limit='10GB', local_directory="/data3/epyc/projects3/sean_hipscat", dashboard_address=':39948', threads_per_worker=1)
client
```

```python
gaia = lsdb.read_hipscat("/data3/epyc/projects3/ivoa_demo/gaia/catalog")
gaia
```

```python
gaia_cone = gaia.cone_search(10,10, 0.1)
gaia_cone.compute()
```

```python
ztf = lsdb.read_hipscat("/data3/epyc/data3/hipscat/catalogs/ztf_axs/ztf_dr14")
ztf
```

```python
ztf_x_gaia_cone = ztf.crossmatch(gaia_cone)
ztf_x_gaia_cone.compute()
```

```python
ztf_sources = lsdb.read_hipscat("/data3/epyc/data3/hipscat/catalogs/ztf_axs/ztf_source")
ztf_sources
```

```python
ztf_x_gaia_cone.join(ztf_sources, left_on="ps1_objid_ztf_dr14", right_on="ps1_objid").compute()
```
